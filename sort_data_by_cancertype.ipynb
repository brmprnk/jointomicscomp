```python
rna = pd.read_csv("/Users/bram/Desktop/CSE3000/data/common/clamped_3modal_RNASeqV2_3000MAD_COMMON.csv", index_col=0)
```

```python
common = []
cancer_patients = info.index.values
cancer_types = []

# Get all occurences of the patient in the original data file.
for patient in cancer_patients:
        x = rna[[patient in label for label in rna.index]]
        if x.empty:
            continue

        common.append(x)
        for _ in range(len(x)):
            cancer_types.append(info.loc[patient]["cancer type abbreviation"])
            
# common now holds all the rows of the dataframe, ordered by patients with known cancertypes
common
```

```python
rna_bycancertype = pd.concat(common)
rna_bycancertype
```

```python
gcn = pd.read_csv("/Users/bram/Desktop/CSE3000/data/common/clamped_3modal_GCN_Gistic2_3000MAD_COMMON.csv", index_col=0)
gcn_bycancertype = gcn.loc[rna_bycancertype.index.values]
dna = pd.read_csv("/Users/bram/Desktop/CSE3000/data/common/clamped_3modal_DNAMethylation_3000MAD_COMMON.csv", index_col=0)
dna_bycancertype = dna.loc[rna_bycancertype.index.values]
```

```python
# Write to output files
rna_bycancertype.to_csv("/Users/bram/Desktop/CSE3000/data/clamped_3modal_RNA_3000MAD_cancertypeknown.csv")
gcn_bycancertype.to_csv("/Users/bram/Desktop/CSE3000/data/clamped_3modal_GCN_3000MAD_cancertypeknown.csv")
dna_bycancertype.to_csv("/Users/bram/Desktop/CSE3000/data/clamped_3modal_DNA_3000MAD_cancertypeknown.csv")
```
